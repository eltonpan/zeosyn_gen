```python
import pickle

# Load data from a pickle file
with open('../data/241002_k222_etc_mols_osda_priors_0.pkl', 'rb') as file:
    df = pickle.load(file)

# Print the loaded data
df.columns
```

```python
df
```

```python
cols = ['asphericity', 
        # 'axes', 
        'formal_charge', 'free_sasa', 'mol_weight', 'npr1', 'npr2', 'num_rot_bonds', 'pmi1', 'pmi2', 'pmi3', 'spherocity_index', 'mol_volume']
```

```python
df_cleaned = df[cols]
```

```python
df_cleaned.columns = [x+'_mean_0' if x not in ['formal_charge', 'mol_weight'] else x for x in df_cleaned.columns]
df_cleaned.insert(loc=1, column='axes_mean_0', value=[x[0] for x in df['axes']])
df_cleaned.insert(loc=2, column='axes_mean_1', value=[x[1] for x in df['axes']])
df_cleaned = df_cleaned.rename(columns={'mol_volume_mean_0': 'volume_mean_0'})
df_cleaned.insert(loc=0, column='osda smiles', value=df_cleaned.index)
df_cleaned.reset_index(drop=True, inplace=True)
df_cleaned
```
