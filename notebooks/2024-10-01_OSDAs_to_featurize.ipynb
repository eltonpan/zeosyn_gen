```python
from rdkit import Chem

# Note: These have been cleaned
smiles = [
    'C1COCCN2CCOCCOCCN(CCO1)CCOCCOCC2', # K222
    'C1CC[N+]2(C1)CCCC2', # OSDA 10 from https://www.science.org/doi/epdf/10.1126/science.abh3350
    'C1CC[N+]2(CC1)CCCCC2', # OSDA 11 from https://www.science.org/doi/epdf/10.1126/science.abh3350
    'C1CCC[N+]2(CC1)CCCCC2', # OSDA 12 from https://www.science.org/doi/epdf/10.1126/science.abh3350
    'C1CCC[N+]2(CC1)CCCCCC2', # OSDA 13 from https://www.science.org/doi/epdf/10.1126/science.abh3350
]

mols = [Chem.MolFromSmiles(s) for s in smiles]
Chem.Draw.MolsToGridImage(mols, molsPerRow=4, subImgSize=(200, 200), legends=smiles)
```
