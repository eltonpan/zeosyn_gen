```python
unique_precs = []
for prec_set in precs:
    if type(prec_set) == str:
        prec_set = prec_set.split(', ')

        for prec in prec_set:
            unique_precs.append(prec)
unique_precs = set(unique_precs)
```

```python
'ludox' in ' Ludox AS-40'.lower()
```

```python
prec_dict = {}
for i ,prec in enumerate(unique_precs):
    prec_class = []
    prec_common_name = None
    prec_lower = prec.lower()
    if 'ludox' in prec_lower or 'silica' in prec_lower or 'silicon' in prec_lower or 'Si' in prec or 'silane' in prec_lower:
        prec_class.append('Si')
    if 'aluminum' in prec_lower or 'aluminium' in prec_lower or 'Al' in prec or 'alumina' in prec_lower:
        prec_class.append('Al')
    if 'germanium' in prec_lower or 'Ge' in prec:
        prec_class.append('Ge')
    if 'boro' in prec_lower or 'B' in prec:
        prec_class.append('B')
    if 'sodium' in prec_lower or 'Na' in prec:
        prec_class.append('Na')
    if 'potassium' in prec_lower or 'K' in prec:
        prec_class.append('K')
    if 'hydroxide' in prec_lower or 'OH' in prec:
        prec_class.append('OH')
    if 'fluoride' in prec_lower or 'F' in prec:
        prec_class.append('F')
    
    prec_dict[prec] = {'class(es):': prec_class, 'common name': prec_common_name}
    print(i, prec)

# with open(f"prec_rec/prec_dict.json", "w") as outfile:
#     json.dump(prec_dict, outfile, indent=4)
    
```
